```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig,ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize= (5,3))
ax.bar(x,y)
```

```python
fig,ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title ('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
           marker = '*',
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = 's',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')

ax.grid()
```
